```python
import json
import pandas as pd
from glob import glob

# Find all the results.json files in the arxiv_data/runs/ folder
json_files = glob("src/output/leverco_shortened/runs/*/results.json")

# Create an empty list to store the data from all the files
data = []

# Loop through each file and read the data
for file in json_files:
    with open(file, "r") as f:
        data.append(json.load(f))
        
print(len(json_files))

df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
df2 = df[df['attempts'] != -1]
```
